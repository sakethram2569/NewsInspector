# Fake News Detection

```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import re
import string
```

```python
data_fake = pd.read_csv('Fake.csv')
data_true = pd.read_csv('True.csv')
```

```python
data_fake.head()
```

```python
data_true.head()
```

```python
data_fake["class"] = 0
data_true["class"] = 1
```

```python
data_fake.shape, data_true.shape
```

```python
data_fake_manual_testing = data_fake.tail(50)
for i in range(23480, 23430, -1):
    data_fake.drop([i], axis = 0, inplace = True)

data_true_manual_testing = data_true.tail(50)
for i in range(21416, 21366, -1):
    data_true.drop([i], axis = 0, inplace = True)
```

```python
data_fake.shape, data_true.shape
```

```python
data_fake_manual_testing.shape, data_true_manual_testing.shape
```

```python
data_fake_manual_testing.head(10)
```

```python
data_true_manual_testing.head(10)
```

```python
data_merge = pd.concat([data_true, data_fake], axis = 0)
```

```python
data_merge.shape
```

```python
data_merge.head(10)
```

```python
data_merge.columns
```

```python
data_train = data_merge.drop(["title", 'subject', "date"], axis = 1)
```

```python
data_train.columns
```

```python
data_train.isnull().sum()
```

```python
data_train = data_train.sample(frac = 1)
```

```python
data_train.head(10)
```

```python
data_train.reset_index(inplace = True)
data_train.drop(['index'], axis = 1, inplace = True)
```

```python
data_train.head(10)
```

```python
def wordopt (text):
    text = text.lower()
    text = re.sub(r'\[.*?\]', '', text)
    text = re.sub(r'\W', '', text)
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'<.*?>+', '', text)
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub(r'\n', '', text)
    text = re.sub(r'\w*\d\w*', '', text)
    return text
```

```python
 data_train['text'] = data_train['text'].apply(wordopt)
```

```python
x = data_train['text']
y = data_train['class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR.fit(xv_train, y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier()
DT.fit(xv_train, y_train)
```

```python
pred_dt = DT.predict(xv_test)
```

```python
DT.score(xv_test, y_test)
```
